**Box Plot**

**Box Plot of House Prices**

The box plot represents the distribution of house prices in the Delhi Housing dataset.

It helps to understand:

Median Price – The central value of property prices

Interquartile Range (IQR) – The spread of the middle 50% of house prices

Minimum & Maximum Prices – Overall pricing range

Outliers – Properties with unusually high or low prices

From the visualization, most properties fall within a specific price range, while a few premium or low-valued houses appear as outliers.

```python
# Box Plot of House Prices

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.boxplot(y=dataset['Price'], color='skyblue')

plt.title('Box Plot of House Prices')
plt.ylabel('Price')
plt.show()
```

Most house prices are concentrated within a specific mid-range (middle 50% of data).

The median price lies near the center of the box, indicating balanced price distribution.

The spread (IQR) shows moderate variation in property prices.

A few properties appear as outliers, representing either luxury homes (very high price) or unusually low-priced properties.

Overall, the dataset shows some price variability but is not extremely skewed.

```python
total_houses = dataset.shape[0]
avg_price = dataset['Price'].mean()
max_price = dataset['Price'].max()

latest_year = dataset['YearBuilt'].max()
latest_avg_price = dataset[dataset['YearBuilt']==latest_year]['Price'].mean()
print("Total Houses:", total_houses)
print("Average Price:", avg_price)
print("Highest Price:", max_price)
print("Latest Year Avg Price:", latest_avg_price)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(18,8))

fig.text(0.05, 0.92, f"Total Houses\n{total_houses/1000:.0f}K",
         fontsize=16, bbox=dict(facecolor='lightgray'))

fig.text(0.25, 0.92, f"Average Price\n{avg_price/1000:.2f}K",
         fontsize=16, bbox=dict(facecolor='lightgray'))

fig.text(0.45, 0.92, f"Highest Price\n{max_price/1000:.2f}K",
         fontsize=16, bbox=dict(facecolor='lightgray'))

fig.text(0.65, 0.92, f"Latest Year Avg Price\n{latest_avg_price/1000:.2f}K",
         fontsize=16, bbox=dict(facecolor='lightgray'))

plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Neighborhood', y='Price', data=dataset, estimator=np.mean)
plt.title("Average Price by Neighborhood")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Bedrooms', y='Price', hue='Neighborhood',
            data=dataset, estimator=np.mean)

plt.title("Average Price by Bedrooms & Neighborhood")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Bathrooms', y='Price', hue='Neighborhood',
            data=dataset, estimator=np.mean)

plt.title("Average Price by Bathrooms & Neighborhood")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(x='YearBuilt', y='Price', data=dataset, estimator=np.mean)
plt.title("Average Price by Year Built")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='SquareFeet', y='Price', data=dataset)
plt.title("Price vs Square Feet")
plt.show()
```

```python
neigh_count = dataset['Neighborhood'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(neigh_count,
        labels=neigh_count.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title("House Distribution by Neighborhood")
plt.show()
```

```python
bath_count = dataset['Bathrooms'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(bath_count,
        labels=bath_count.index,
        autopct='%1.1f%%',
        startangle=90)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Bathroom Distribution")
plt.show()
```

```python
bath_count = dataset['Bathrooms'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(bath_count,
        labels=bath_count.index,
        autopct='%1.1f%%',
        startangle=90)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Bathroom Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.hist(dataset['Price'], bins=20)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,5))

numeric_data = dataset.select_dtypes(include=['int64','float64'])

sns.heatmap(numeric_data.corr(), annot=True)

plt.title("Correlation Heatmap")
plt.show()
```
